```python

# from the main table    
first_names = []    
last_names = []
all_scores = []
event_21_1 = []
event_21_2 = []
event_21_3 = []
event_21_4 = []

# from the profile 
profile_age = []
profile_country = []
profile_hight = []
profile_weight = []
year = []
division = []
rank_worldwide = []
rank_by_region = []
rank_by_country = []
profile_Back_Squat = []
profile_Clean_and_Jerk = []
profile_Snatch = []
profile_Deadlift = []
profile_Fight_Gone_Bad = []
profile_Max_Pull_ups = []
profile_Fran = []
profile_Grace = []
profile_Helen = []
profile_Filthy = []
profile_Sprint = []
profile_Run_5k = []

s = Service(ChromeDriverManager().install())
print("len of main table links:",len(mans_open_main_table_page_links))#deleteme
checkout = 1#deleteme        
for mans_open_main_table_page_link in mans_open_main_table_page_links[67:]:#if want can use -> from where to where inside the array of links[3:4]
    print("item:",checkout,"outside mans_open_main_table_page_links loop")#deleteme
    checkout+=1 #deleteme
    options = Options()
    options.headless = True
    options.add_argument("--window-size=1920,1200")
    
    driver = webdriver.Chrome(options=options, service=s)
    driver.maximize_window()
    driver.get(mans_open_main_table_page_link)
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#leaderboardSponsorVisible > div > table > tbody > tr:nth-child(1) > td.name > div > div.top > div.full-name > div.last-name"))
    )
    
    # הפיכת הקובץ מסלניום לביטויופול-סוופ
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # לקיחת כל השמות הפרטיים מהדף הראשי
    for div in soup.find_all("div", attrs={"class":"first-name"}):
        first_names.append(div.get_text().strip())
        
    # לקיחת כל השמות משפחה מהדף הראשי
    for div in soup.find_all("div", attrs={"class":"last-name"}):
        last_names.append(div.get_text().strip())
    
    # נכנס לתוך ה'דיב' הנכון ולוקח את כל התוצאות של כל השחקנים מהטבלה הראשית
    for tr in soup.tbody:
        for td in tr.select("tr > td")[2]:
            all_scores.append(td.get_text())
            
    # לוקח את כל התוצאות ב4 התחרויות של כל שחקן - זאת אומרת שיש 4*50 שחקנים = סה"כ 200 תוצאות
    #READ ME - every 4 rows its one row/ object.
    all_events_temp = []
    for div in soup.find_all("div", attrs={"class":"rank-result"}):
        all_events_temp.append(div.get_text().strip())
    length = len(all_events_temp)
    index_eventes_for_crawling = 0
    while index_eventes_for_crawling < length:
        event_21_1.append(all_events_temp[index_eventes_for_crawling])
        event_21_2.append(all_events_temp[index_eventes_for_crawling+1])
        event_21_3.append(all_events_temp[index_eventes_for_crawling+2])
        event_21_4.append(all_events_temp[index_eventes_for_crawling+3])
        index_eventes_for_crawling+=4
        
    # אוסף לתוך המערך 'פרופיל לינקס' את כל הלינקים לדפי השחקנים מאותו עמוד טבלה ראשי
    profile_links = []### this array need to changee every page on the main table pages !!!
    for a in soup.find_all("a", attrs={"class": "profile-link"}):
        profile_links.append("https://games.crossfit.com"+a.get('href'))
    
    driver.quit() # free -> driver -> from outside loop
    
    
    #    מעבר על כל העמודים האישיים של השחקנים בתוך העמוד טבלה הנוכחי !
    # איחוד הקרואלינג של 50 עמודי השחקנים לטובת מעבר על כל הלינקים של כל המשתמשים 
    # סלניום על הדף של הפרופיל האישי של השקחן והפיכת הדף האישי לאובייקט מסוג ביוטיופול-סוופ
    print("len of profile links:",len(profile_links))#deleteme
    checkinside = 1#deleteme
    for profile_link in profile_links:         
        print("item:",checkinside,"inside profile links loop")#deleteme
        checkinside+=1 #deleteme
        options = Options()
        options.headless = True
        options.add_argument("--window-size=1920,1200")
        
        driver = webdriver.Chrome(options=options, service=s)
        driver.maximize_window()
        driver.get(profile_link)
        #    element = WebDriverWait(driver, 10).until(
        #        EC.presence_of_element_located((By.CSS_SELECTOR, "#leaderboardSponsorVisible > div > table > tbody > tr:nth-child(1) > td.name > div > div.top > div.full-name > div.last-name"))
        #    )
        
        soup_profile_link = BeautifulSoup(driver.page_source, 'html.parser')
        # print(soup_profile_link.prettify())
        
        #קודם נבדוק אם העמוד קיים ולא ריק מתוכן
        #במידה ולא קיים  -> מכניס לכל ערכי הפרופיל לינק את הערך 'כלום '
        if(soup_profile_link.find("div", attrs = {"class":"no-rep"})):
            profile_age.append(None)
            profile_country.append(None)
            profile_hight.append(None)
            profile_weight.append(None)
            profile_Back_Squat.append(None)
            profile_Clean_and_Jerk.append(None)
            profile_Snatch.append(None)
            profile_Deadlift.append(None)
            profile_Fight_Gone_Bad.append(None)
            profile_Max_Pull_ups.append(None)
            profile_Fran.append(None)
            profile_Grace.append(None)
            profile_Helen.append(None)
            profile_Filthy.append(None)
            profile_Sprint.append(None)
            profile_Run_5k.append(None)
            continue
                
        # לקיחת הגיל של השחקן מהעמוד האישי שלו
        profile_age.append(soup_profile_link.find("div", attrs = {"class":"text lg"}).get_text(strip = True))
        
        # לקיחת המדינה של השחקן מהעמוד האישי שלו
        profile_country.append(soup_profile_link.find("div", attrs = {"class":"text"}).get_text(strip = True))
        
        # לקיחת הגובה והמשקל של השחקן מהעמוד האישי שלו והכנסתם למשתנים
        temp_hight_and_weight = [div.get_text().strip() for div in soup_profile_link.find_all("div", attrs={"class":"text lg nowrap"})]
        profile_hight.append(temp_hight_and_weight[0])
        profile_weight.append(temp_hight_and_weight[1])
    
#         # לוקח את כל טבלת ה'תחרויות אוופן' של השחקן מהעמוד האישי שלו, ומפצל אותה למערכים שישמשו עמודות בטבלה הסופית
#         # בעבר  'open' נתונים אישיים מתחרות ה
#         temp_table = [div.get_text().strip() for div in soup_profile_link.tbody.find_all("td", attrs={"class":"cell"})]
#         length = len(temp_table)
#         index_on_player_profile_table = 0
#         while index_on_player_profile_table < length:
#             year.append(temp_table[index_on_player_profile_table])
#             division.append(temp_table[index_on_player_profile_table+1])
#             rank_worldwide.append(temp_table[index_on_player_profile_table+2])
#             rank_by_region.append(temp_table[index_on_player_profile_table+3])
#             rank_by_country.append(temp_table[index_on_player_profile_table+4])
#             index_on_player_profile_table+=5
    
        # לוקח את כל טבלת הסטאטיסטיקות 'סטאטס' שבתחתית העמוד של השחקן - > ומכניס את המידע למשתנים המתאימים
        soup_profile_link_stats = soup_profile_link.find("ul", attrs={"class":"stats-container"})
        profile_stats_1 = []
        for index_on_player_profile_stats in soup_profile_link_stats.find_all("td"):
            profile_stats_1.append(index_on_player_profile_stats.get_text().strip())
        profile_Back_Squat.append(profile_stats_1[0])
        profile_Clean_and_Jerk.append(profile_stats_1[1])
        profile_Snatch.append(profile_stats_1[2])
        profile_Deadlift.append(profile_stats_1[3])
        profile_Fight_Gone_Bad.append(profile_stats_1[4])
        profile_Max_Pull_ups.append(profile_stats_1[5])
        profile_Fran.append(profile_stats_1[6])
        profile_Grace.append(profile_stats_1[7])
        profile_Helen.append(profile_stats_1[8])
        profile_Filthy.append(profile_stats_1[9])
        profile_Sprint.append(profile_stats_1[10])
        profile_Run_5k.append(profile_stats_1[11])
        
        driver.quit() # free -> driver -> from inside loop
print("Done !")        
```

```python
df.drop('Country', axis=1, inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)


#split events
df['Event 21.1'].str.split(',', expand=True)
df[['Event_1_score', 'Event_1_time']] = df['Event 21.1'].str.split(' ',n=1, expand=True)

df['Event 21.2'].str.split(',', expand=True)
df[['Event_2_score', 'Event_2_time']] = df['Event 21.2'].str.split(' ',n=1, expand=True)

df['Event 21.3'].str.split(',', expand=True)
df[['Event_3_score', 'Event_3_time']] = df['Event 21.3'].str.split(' ',n=1, expand=True)

df['Event 21.4'].str.split(',', expand=True)
df[['Event_4_score', 'Event_4_lbs']] = df['Event 21.4'].str.split(' ',n=1, expand=True)


df['Snatch'].str.split(' ', expand=True)
df[['Snatch', 'kg-lb']] = df['Snatch'].str.split(' ',n=1, expand=True)
df['Snatch'] = pd.to_numeric(df['Snatch'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)
df['Snatch'] = df['Snatch'].astype(int)
df.loc[df['kg-lb']=='lb', ['Snatch']] *= 0.45

df = df.replace(0, np.nan, regex=True)
df.fillna(value=df['Snatch'].mean() ,inplace =True)   

df['Deadlift'].str.split(' ', expand=True)
df[['Deadlift', 'kg-lb']] = df['Deadlift'].str.split(' ',n=1, expand=True)

df['Deadlift'] = pd.to_numeric(df['Deadlift'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)
df['Deadlift'] = df['Deadlift'].astype(int)
df.loc[df['kg-lb']=='lb', ['Deadlift']] *= 0.45

df = df.replace(0, np.nan, regex=True)
df.fillna(value=df['Deadlift'].mean() ,inplace =True)   


df['Clean&Jerk'].str.split(' ', expand=True)
df[['Clean&Jerk', 'kg-lb']] = df['Clean&Jerk'].str.split(' ',n=1, expand=True)

df['Clean&Jerk'] = pd.to_numeric(df['Clean&Jerk'], errors='coerce')
df = df.replace(np.nan, 0, regex=True)
df['Clean&Jerk'] = df['Clean&Jerk'].astype(int)

df.loc[df['kg-lb']=='lb', ['Clean&Jerk']] *= 0.45


df = df.replace(0, np.nan, regex=True)
df.fillna(value=df['Clean&Jerk'].mean() ,inplace =True)   


df['Back Squat'].str.split(' ', expand=True)

df[['Back Squat', 'kg-lb']] = df['Back Squat'].str.split(' ',n=1, expand=True)

df['Back Squat'] = pd.to_numeric(df['Back Squat'], errors='coerce')

df = df.replace(np.nan, 0, regex=True)

df['Back Squat'] = df['Back Squat'].astype(int)

df.loc[df['kg-lb']=='lb', ['Back Squat']] *= 0.45


df = df.replace(0, np.nan, regex=True)
df.fillna(value=df['Back Squat'].mean() ,inplace =True)   


df['Weight'].str.split(' ', expand=True)

df[['Weight', 'kg-lb']] = df['Weight'].str.split(' ',n=1, expand=True)


df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')

df = df.replace(np.nan, 0, regex=True)

df['Weight'] = df['Weight'].astype(int)

df.loc[df['kg-lb']=='lb', ['Weight']] *= 0.45

df = df.replace(0, np.nan, regex=True)
df.fillna(value=df['Weight'].mean() ,inplace =True)   

# drop col:

col= ['Fight Gone Bad','Max Pull Ups',  'Fran' ,'Grace', 'Helen' ,'Filthy' ,'Sprint', 'Run 5K']
for x in col:
    df.drop([x], axis=1, inplace=True)
```
